```python
from PIL import Image
import random
# 이우원의 테스트
```

```python
background = Image.open('sky.jpg')
```

```python
imglst=[]
for i in range(1,8):
    img = Image.open(str(i) + ".jpg")
    imglst.append(img)
```

```python
b = 0
for i in range(len(imglst)):
    flag = 0
    if imglst[i].size[0] > imglst[i].size[1]:
        imglst[i] = imglst[i].transpose(Image.ROTATE_90)
        flag = 1
    if i == 0 :
        basewidth = int(background.width * 0.6)
    else:
        basewidth = int(b * 0.9)
    wpercent = (basewidth / float(imglst[i].size[0]))
    hsize = int((float(imglst[i].size[1]) * float(wpercent)))
    imglst[i] = imglst[i].resize((basewidth, hsize), Image.ANTIALIAS)
    b = basewidth
    if flag == 1:
        imglst[i] = imglst[i].transpose(Image.ROTATE_270)
```

```python
back = background.copy()
for img in imglst:
    w = random.randint(0,(background.width - img.width))
    h = random.randint(0,(background.height - img.height))
    print((w,h))
    back.paste(img, (w,h))
```
